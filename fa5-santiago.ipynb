```python
import turtle
import random
import math

wn = turtle.Screen()
wn.bgcolor("lightgreen")

t = turtle.Turtle()
t.shape("turtle")
t.color("blue")
t.speed(0)

def DrawFence(r):
    t.penup()
    t.goto(0, -r)
    t.pendown()
    t.circle(r)

def isOutside(t, r):
    x, y = t.position()
    distance = math.sqrt(x ** 2 + y ** 2)
    return distance > r

def Bounce_off_fence(t, r):
    while isOutside(t, r):
        t.left(180)
        t.forward(10)

radius = 150
DrawFence(radius)

def MoveRandomly():
    t.left(random.randint(0, 360))
    t.forward(random.randint(50, 200))

while True:
    MoveRandomly()
    Bounce_off_fence(t, radius)

wn.mainloop()
```
